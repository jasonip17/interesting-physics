```python
import sys
sys.path.append('../')
import numpy as np
import matplotlib.pyplot as plt
from module.thermo.BEC import N_BEC
from scipy.constants import atomic_mass, k as kB
```

```python
m = 87 * atomic_mass
N = 100_000
V = 1e-15

BEC_dict = N_BEC(N, m, V, T_min=0)
```

```python
plt.figure(figsize=(9, 6))

plt.plot(BEC_dict['T']/BEC_dict['Tc'], BEC_dict['N0'], label='$N_0$ (Ground State)', color='blue')
plt.plot(BEC_dict['T']/BEC_dict['Tc'], BEC_dict['N1'], label='$N_{ex}$ (Excited States)', color='red')
plt.axvline(1.0, color='gray', linestyle='--', label='$T_c$')

plt.title("BEC Simulation")
plt.xlabel("Temperature ($T / T_c$)")
plt.ylabel("Number of Particles")
plt.xlim(0)
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()

plt.show()
```

```python
BEC_dict = N_BEC(N, m, V, T_min=0, T_max=3.979148732593932e-07 * 20, n_points=500)

plt.figure(figsize=(9, 6))
plt.plot(BEC_dict['T']/BEC_dict['Tc'], BEC_dict['z'], color='blue')
plt.axvline(1.0, color='gray', linestyle='--', label='$T_c$')
plt.title("Fugacity")
plt.xlabel("Temperature ($T / T_c$)")
plt.ylabel("z")
plt.xlim(0)
plt.grid(True, alpha=0.3)
plt.ylim(0)
plt.tight_layout()

plt.show()
```

```python
mu = kB * BEC_dict['T'] * np.log(BEC_dict['z'])

plt.figure(figsize=(9, 6))
plt.plot(BEC_dict['T']/BEC_dict['Tc'], mu, color='blue')
plt.axvline(1.0, color='gray', linestyle='--', label='$T_c$')
plt.title('Chemical Potential')
plt.xlabel("Temperature ($T / T_c$)")
plt.ylabel(r'$\mu$')
plt.xlim(0)
plt.grid(True, alpha=0.3)
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(9, 6))
plt.plot(BEC_dict['T']/BEC_dict['Tc'], BEC_dict['Cv'], color='blue')
plt.axvline(1.0, color='gray', linestyle='--', label='$T_c$')
plt.title('Heat Capacity')
plt.xlabel("Temperature ($T / T_c$)")
plt.ylabel(r'$C_v$')
plt.xlim(0,7.5)
plt.grid(True, alpha=0.3)
plt.tight_layout()

plt.show()
```
